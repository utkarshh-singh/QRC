```python
import numpy as np
from ESN2 import EchoStateNetwork
```

```python
from reservoirs import CPRC, GBPermanents, GBSampling
from utility import Evaluator, BSNetwork
from properties import lyapunov_exponent, separation_property, washout_test, memory_capacity
```

```python
BSN = BSNetwork(q_modes=3, network_type="Custom", reps =2) #[[3,2],[2,1],[3,2], [3,2],[2,1],[3,2]]
network_connections = BSN.get_network()
BSN.get_info()
```

```python
dim = 4
X = np.random.randn(200, dim)
```

```python
res = GBPermanents(network_connections=network_connections, classical=True, kernel= False, n_jobs=8)
esn1 = EchoStateNetwork(
    reservoir=res, 
    dim=dim, 
    regularization=1e-6, 
    alpha=0.3, 
    show_progress=True, 
    approach='feedback', # 'feedback', or 'time-multiplexing'
    model_type='ridge',  # Options: 'ridge', 'lasso', 'linear', 'svr', 'svc'
    limit = 0.8)
```

```python
reservoir = esn1

lyapunov_exp = lyapunov_exponent(reservoir, X)
print(f"Lyapunov Exponent: {lyapunov_exp}")

# Compute separation property
X1, X2 = X + 0.01, X - 0.01  # Slightly different inputs
separation = separation_property(reservoir, X1, X2)
print(f"Separation Property: {separation}")

# Perform washout test
washout = washout_test(reservoir, X)
print(f"Washout Influence: {washout}")

# Compute memory capacity
memory_capacity_ = memory_capacity(reservoir, X)
print(f"Memory Capacity: {memory_capacity_}")
```

```python
res2 = CPRC(dim=dim, execution_mode='simulation')
esn2 = EchoStateNetwork(
    reservoir=res2, 
    dim=dim, 
    regularization=1e-6, 
    alpha=1, 
    show_progress=True, 
    approach='average', # 'feedback', or 'time-multiplexing', 'average', 'direct', 'concatenation'
    model_type='ridge',  # Options: 'ridge', 'lasso', 'linear', 'svr', 'svc'
    limit = 0.6)
```

```python
reservoir = esn2
lyapunov_exp = lyapunov_exponent(reservoir, X)
print(f"Lyapunov Exponent: {lyapunov_exp}")

# Compute separation property
X1, X2 = X + 0.01, X - 0.01  # Slightly different inputs
separation = separation_property(reservoir, X1, X2)
print(f"Separation Property: {separation}")

# Perform washout test
washout = washout_test(reservoir, X)
print(f"Washout Influence: {washout}")

# Compute memory capacity
memory_capacity = memory_capacity(reservoir, X)
print(f"Memory Capacity: {memory_capacity}")
```
