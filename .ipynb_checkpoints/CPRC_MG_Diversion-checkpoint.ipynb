```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.special import expit as sigmoid
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge, LinearRegression
```

```python
from utils import DataHandler, plot_closed_loop_forecast, closed_loop_predict_from_X_test
from ESN import ESNetwork, ClassicalRC
# from ESN2 import EchoStateNetwork
from reservoirs import CPRC, GBPermanents
```

```python
window_size = 20
n_samples = 2000
# X, Y = DataHandler().load_dataset('narma', n_samples = 1000, window_size=4, plot=True)
X, Y = DataHandler().load_dataset('mackey_glass', n_samples = n_samples, tau = 12, window_size=window_size,prediction_horizon=1, plot=True)
```

```python
tr = 0.5
train_size = int(n_samples*tr)  # Training set size
# Split into training and testing sets
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = Y[:train_size], Y[train_size:]
```

```python
X_test[2]
```

```python
dim=len(X_train[0])
cprc = CPRC(dim=dim, execution_mode='simulation', kernel = True)
esn = ESNetwork(
    reservoir=cprc, 
    dim=dim, 
    regularization=1e-6, 
    alpha=1, 
    show_progress=True, 
    approach='feedback', # 'feedback', or 'time-multiplexing'
    model_type='ridge',  # Options: 'ridge', 'lasso', 'linear', 'svr', 'svc'
    limit = 0.6,
    cpk = True
)
```

```python
esn.fit(X_train, y_train)
```

```python
predictions = esn.predict(X_test)
# print(predictions)
```

```python
rmse = mean_squared_error(y_test, predictions)
print(f"Quantum ESN RMSE: {rmse:.4f}")
a = 0
b = 50
# Plot Predictions
plt.plot(y_test[a:a+b], label="True")
plt.plot(predictions[a:a+b], label="Quantum ESN Pred")
plt.title("Quantum ESN Predictions with Optical Reservoir Computing")
plt.legend()
plt.show()
```

```python
X_test[4], y_test[4:14], predictions[4:14]
```

```python
def rmse_per_timestep(y_true, y_pred):
    return np.sqrt((y_true - y_pred)**2)

m = None   # Closed-loop mode
h = 200
X_test_subset = X_test[:h]
y_test_subset = y_test[:h]

preds = closed_loop_predict_from_X_test(esn, X_test_subset, m=m, steps=h)
rmse_t = rmse_per_timestep(y_test_subset[:len(preds)], preds)

# Plot RMSE over time
plt.figure(figsize=(10, 5))
plt.plot(rmse_t, label=f"m={m}", color='red')
plt.xlabel("Time step")
plt.ylabel("RMSE")
plt.title("RMSE over time (Closed-loop)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
h = 100  
m_vals = [2, 5, 8, 10, 15, 18, 22, 28, 32, 35]
mse_fp = []

for m in m_vals:
    X_test_subset = X_test[:h]
    y_test_subset = y_test[:h]
    
    preds = closed_loop_predict_from_X_test(esn, X_test_subset, m=m, steps=h)
    mse = mean_squared_error(preds, y_test_subset)
    mse_fp.append(mse)
    print(f"m = {m}: MSE = {mse:.6f}")

# Plot MSE vs m
plt.figure(figsize=(8, 5))
plt.plot(m_vals, mse_fp, marker='o')
plt.xlabel("Reset interval (m)")
plt.ylabel("MSE over 200 steps")
plt.title("MSE vs. Reset Interval (m)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
