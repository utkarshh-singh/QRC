```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.special import expit as sigmoid
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge, LinearRegression
```

```python
from utils import DataHandler
from ESN import EchoStateNetwork, ESNetwork
from reservoirs import CPRC, GBPermanents
from properties import verify_ESP, memory_capacity
```

```python
# X, Y = DataHandler().load_dataset('narma', n_samples = 1000, window_size=4, plot=True)
X, Y, Y_ = DataHandler().load_dataset('mixed_waves',n=10, n_samples = 1000, window_size=10, prediction_horizon=5, plot=False, seed = 58)
```

```python
train_size = 750  # Training set size
window_size = 10
# Split into training and testing sets
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = Y[:train_size], Y[train_size:]
```

```python
# X_test[2]
```

```python
dim=len(X_train[0])
cprc = CPRC(dim=dim, execution_mode='simulation', kernel = False)
esn = ESNetwork(
    reservoir=cprc, 
    dim=dim, 
    regularization=1e-6, 
    alpha=1, 
    show_progress=True, 
    approach='time-multiplexing', # 'feedback', or 'time-multiplexing'
    model_type='ridge',  # Options: 'ridge', 'lasso', 'linear', 'svr', 'svc'
    limit = 0.6,
    cpk = False
)
```

```python
esn.fit(X_train, y_train)
```

```python
predictions = esn.predict(X_test)
# print(predictions)
```

```python
rmse = mean_squared_error(y_test, predictions)
print(f"Quantum ESN RMSE: {rmse:.4f}")
a = 0
b = 397
# Plot Predictions
plt.plot(y_test[a:a+b], label="True")
plt.plot(predictions[a:a+b], label="Quantum ESN Pred")
plt.title("Quantum ESN Predictions with Optical Reservoir Computing")
plt.legend()
plt.show()
```

```python
n = 10 
m = 5 
initial_input = X_train[-1]  

future_predictions = esn.predict(X = None, X_test=X_test, initial_input=initial_input, n=n, m=m)
print(future_predictions)
```

```python
states_normal = esn.get_saved_states()
X_train_perturbed = X_train + np.random.normal(0, 0.01, X_train.shape)
esn.fit(X_train_perturbed, y_train)
states_perturbed = esn.get_saved_states()
```

```python
esp_satisfied = verify_ESP(states_normal, states_perturbed)
print(f"ESP Satisfied: {esp_satisfied}")
```

```python
memory_capacity_value = memory_capacity(states_normal, X_train[:, 0])
print(f"Memory Capacity: {memory_capacity_value:.4f}")
```

```python
models = {
    "LinearRegression": LinearRegression(),
    "RidgeRegression": Ridge(alpha=1.0)
}

results = {}
n_predictions = 50  # Number of future values to predict

for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    results[name] = rmse
    
    # Predict next n values iteratively
    future_predictions = list(X_test[-1])
    for _ in range(n_predictions):
        next_input = np.array(future_predictions[-window_size:]).reshape(1, -1)
        next_pred = model.predict(next_input)[0]
        future_predictions.append(next_pred)
        
    predict = model.predict(X_test)
    plt.figure(figsize=(10, 4))
    plt.plot(y_test[0:100], label="True")
    plt.plot(predict[0:100], label="predicted")
#     plt.plot(predictions, label=f"{name} Pred")
#     plt.axvline(len(y_test), color='r', linestyle='--', label="Future Predictions Start")
#     plt.plot(range(len(y_test), len(y_test) + n_predictions), future_predictions[-n_predictions:], label=f"{name} Future Pred")
    plt.title(f"{name} Predictions on NARMA10")
    plt.legend()
    plt.show()

# Compare results
print("Model RMSE Scores:")
for model, rmse in results.items():
    print(f"{model}: {rmse:.4f}")
```
