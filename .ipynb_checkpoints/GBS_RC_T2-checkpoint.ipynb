```python
import numpy as np
from tqdm import tqdm
from sklearn.linear_model import Ridge

class GBEchoStateNetwork:
    def __init__(self, dim=4, regularization=1e-6, num_workers=4, parallel=False, alpha=0.8, show_progress=True):
        """
        Gaussian Boson Sampling (GBS) Echo State Network with Feedback.

        Parameters:
            dim (int): Fixed input size for the GBS device (default: 4).
            regularization (float): Ridge regression regularization parameter.
            num_workers (int): Number of workers for parallel processing in GBS.
            parallel (bool): Whether to use parallel computation.
            alpha (float): Feedback memory parameter (0 to 1). Higher values retain more memory.
            show_progress (bool): Whether to display a progress bar.
        """
        self.dim = dim
        self.num_workers = num_workers
        self.parallel = parallel
        self.alpha = alpha  # Memory mixing factor
        self.show_progress = show_progress  # Toggle progress bar
        self.quantum_reservoir = GBSampling()
        self.regularization = regularization
        self.model = None  # Ridge Regression model
        self.prev_output = np.zeros(dim)  # Initialize previous GBS output (feedback memory)

    def fit(self, X, Y):
        """
        Train the reservoir using feedback-enhanced inputs.
        
        Parameters:
            X (array-like): Input time series data (sliding window format).
            Y (array-like): Target values for training.
        """
        quantum_states = []
        prev_output = np.zeros(self.dim)  # Initialize feedback memory

        iterator = tqdm(X, desc="Training Progress", unit=" sample", disable=not self.show_progress)

        for x in iterator:
            x = np.array(x, dtype=np.float32)  # Ensure x is a NumPy array
            prev_output = np.array(prev_output)
            
            # Apply feedback: Combine current input with past output memory
            modified_input = self.alpha * x + (1 - self.alpha) * prev_output

            # Process through the quantum reservoir
            quantum_state = self.quantum_reservoir.Probs(modified_input, num_workers=self.num_workers, parallel=self.parallel)

            quantum_states.append(quantum_state)
            prev_output = quantum_state  # Update feedback memory

        quantum_states = np.array(quantum_states)

        # Ensure shape consistency
        if quantum_states.ndim == 1:
            quantum_states = quantum_states.reshape(-1, 1)

        # Train Ridge Regression model
        self.model = Ridge(alpha=self.regularization).fit(quantum_states, Y)


    def predict(self, X):
        """
        Predict output using the trained Ridge Regression model with feedback.

        Parameters:
            X (array-like): Input time series data.
        
        Returns:
            np.ndarray: Predicted values.
        """
        if self.model is None:
            raise ValueError("Model has not been trained yet. Call `fit` first.")

        quantum_states = []
        prev_output = np.zeros(self.dim)  # Initialize feedback memory

        iterator = tqdm(X, desc="Prediction Progress", unit=" sample", disable=not self.show_progress)

        for x in iterator:
            x = np.array(x, dtype=np.float32)  # Ensure x is a NumPy array
            prev_output = np.array(prev_output)
            
            # Apply feedback
            modified_input = self.alpha * x + (1 - self.alpha) * prev_output

            # Get reservoir transformation
            quantum_state = self.quantum_reservoir.Probs(modified_input, num_workers=self.num_workers, parallel=self.parallel)

            quantum_states.append(quantum_state)
            prev_output = quantum_state  # Update feedback memory

        quantum_states = np.array(quantum_states)

        if quantum_states.ndim == 1:
            quantum_states = quantum_states.reshape(-1, 1)

        return self.model.predict(quantum_states)


    def predict_next_n_trained(self, initial_input, X_test, n, m):
        """
        Predicts the next `n` values while resetting to the true `X_test` values every `m` predictions.

        Parameters:
            initial_input (array-like): The initial input sequence to start predictions.
            X_test (array-like): The original test data to reset after every `m` predictions.
            n (int): Number of future values to predict.
            m (int): Reset interval (after how many predictions to reset to X_test).

        Returns:
            np.ndarray: Array of predicted values.
        """
        if self.model is None:
            raise ValueError("Model has not been trained yet. Call `fit` first.")

        predictions = []
        current_input = np.array(initial_input).reshape(1, -1)
        test_index = 0  # Track X_test reset index
        prev_output = np.zeros(self.dim)  # Initialize previous GBS output for feedback

        iterator = tqdm(range(n), desc="Sequential Prediction Progress", unit=" step", disable=not self.show_progress)

        for i in iterator:
            # Apply feedback mechanism
            modified_input = self.alpha * current_input[0] + (1 - self.alpha) * prev_output

            # Get quantum reservoir output
            quantum_state = np.array([self.quantum_reservoir.Probs(modified_input, 
                                                                   num_workers=self.num_workers, 
                                                                   parallel=self.parallel)])

            if quantum_state.ndim == 1:
                quantum_state = quantum_state.reshape(-1, 1)

            # Predict next value
            next_pred = self.model.predict(quantum_state)
            predictions.append(next_pred.flatten()[0])  # Store the scalar prediction

            # Update previous output for feedback
            prev_output = quantum_state.flatten()

            # Every `m` predictions, reset to the true value from X_test
            if (i + 1) % m == 0 and test_index < len(X_test):
                current_input = np.array(X_test[test_index]).reshape(1, -1)
                test_index += 1
            else:
                # Shift left and insert predicted value
                current_input = np.roll(current_input, -1)
                current_input[0, -1] = next_pred.flatten()[0]

        return np.array(predictions)
```

```python
def generate_narma(n_samples=2000, order=10):
    np.random.seed(42)
    u = np.random.uniform(0, 0.5, n_samples)
    y = np.zeros(n_samples)
    for t in range(order, n_samples - 1):
        y[t + 1] = (0.3 * y[t] + 0.05 * y[t] * np.sum(y[t - order:t]) 
                     + 1.5 * u[t - order] * u[t] + 0.1)
    return u, y
```

```python
def create_sliding_windows(data, window_size=4):
    X = np.array([data[i:i+window_size].flatten() for i in range(len(data) - window_size)])
    return X
```

```python
# Generate NARMA10 dataset
sequence_length = 2000  # Length of time series
train_size = 1500  # Training set size

U, Y = generate_narma(sequence_length)

# Prepare sliding window inputs for GBS (each window = 4 time steps)
X = create_sliding_windows(U, window_size=4)
Y = Y[4:]  # Align targets with windowed inputs

# Split into training and testing sets
X_train, X_test = X[:train_size], X[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]
```

```python
X_train[0]
```

```python
gbs_reservoir = GBEchoStateNetwork(dim=4, regularization=1e-6, parallel=False, alpha=0.7, show_progress=True)
gbs_reservoir.fit(X_train, Y_train)
```

```python
predictions = gbs_reservoir.predict(X_test)
nmse = np.mean((Y_test - predictions) ** 2) / np.var(Y_test)
print(f"NARMA10 NMSE: {nmse:.5f}")
```

```python
# Plot the results
samp=20
plt.figure(figsize=(10, 4))
plt.plot(range(samp), Y_test[0:samp], label="True Output", linestyle="dashed", alpha=0.7)
plt.plot(range(samp), predictions[0:samp], label="GBS Reservoir Prediction", linewidth=2)
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.title("GBS Reservoir Prediction on NARMA10")
plt.legend()
plt.show()
```

```python

def MG_series(N=20000, b=0.1, c=0.3, tau=18, window_size=4, initial_conditions=None, plot=False, return_dataframe=True, min_plot_points=100):
    """
    Generates a Mackey-Glass time series dataset and applies a sliding window for inputs and targets.

    Parameters:
        N (int): Number of data points to generate.
        b (float): Decay rate parameter.
        c (float): Production rate parameter.
        tau (int): Time delay in the system.
        window_size (int): Number of past values to use as input for each prediction.
        initial_conditions (list, optional): Initial values to start the series.
        plot (bool): If True, plots the first 500 points of the dataset.
        return_dataframe (bool): If True, returns a DataFrame. If False, returns X (time series values) and y (shifted targets).

    Returns:
        pd.DataFrame or (X, y): A DataFrame containing input sequences and targets if return_dataframe=True.
                                Otherwise, returns X (input sequences) and y (targets).
    """
    if initial_conditions is None:
        initial_conditions = [
            0.9697, 0.9699, 0.9794, 1.0003, 1.0319, 1.0703, 1.1076, 1.1352,
            1.1485, 1.1482, 1.1383, 1.1234, 1.1072, 1.0928, 1.0820, 1.0756,
            1.0739, 1.0759
        ]

    # Initialize the time series with given initial conditions
    y = initial_conditions.copy()

    # Compute the Mackey-Glass series
    for n in range(17, N + 99):
        y.append(y[n] - b * y[n] + c * y[n - tau] / (1 + y[n - tau] ** 10))

    # Trim the initial transient phase (first 100 data points)
    y = np.array(y[100:])

    # Create input sequences and target values using a sliding window
    X, Y = [], []
    for i in range(len(y) - window_size):
        X.append(y[i : i + window_size])  # Last `window_size` values as input
        Y.append(y[i + window_size])  # Next value as target

    X, Y = np.array(X), np.array(Y) 

    # Convert to DataFrame
    df = pd.DataFrame(X, columns=[f"Lag {i+1}" for i in range(window_size)])
    df["Target"] = Y

    if plot:
        plot_points = min(min_plot_points, len(Y))
        plt.figure(figsize=(8, 3))
        plt.plot(range(plot_points), Y[:plot_points], color='blue', alpha=0.7, label='Mackey-Glass Targets')
        plt.xlabel("Time Step")
        plt.ylabel("Value")
        plt.title(f"Mackey-Glass Time Series Targets (First {plot_points} Points)")
        plt.legend()
        plt.grid(True)
        plt.show()

    if return_dataframe:
        return df
    else:
        return X, Y
```
