```python
def process_counts(counts, n_qubits, key_type="hexa"):
    binary_counts = {}
    for k, v in counts.items():
        if key_type == "hexa":
            binary_key = format(int(k, 16), f'0{n_qubits}b')
        elif key_type == "decimal":
            binary_key = format(int(k), f'0{n_qubits}b')
        elif key_type == "binary":
            binary_key = format(int(k, 2), f'0{n_qubits}b')
        else:
            raise ValueError("key_type must be one of: 'hexa', 'decimal', 'binary'")
        binary_counts[binary_key] = v

    all_bitstrings = [format(i, f'0{n_qubits}b') for i in range(2**n_qubits)]
    full_counts = {b: binary_counts.get(b, 0) for b in all_bitstrings}
    return full_counts
```

```python
jobs = service.jobs(limit=50)
```

```python
# jobs = jobs_[1:-1]
jobs
```

```python
import os
import numpy as np
import pickle
from tqdm import tqdm

num_qubits_ = 6
shots = 1024

# Placeholder for results
reservoir_states = []
job_ids_used = []
errors = []

# Process each job
for job in tqdm(jobs, desc="Processing IBM Kyiv jobs"):
    job_id = job.job_id()
    try:
        result = job.result()
#         print(result)
        counts = result[0].data.meas.get_counts()
        fid = process_counts(counts, n_qubits=num_qubits_, key_type="binary").values()
        state = np.array(list(fid)) / shots

        reservoir_states.append(state)
        job_ids_used.append(job_id)

    except Exception as e:
        print(f"⚠️ Failed for job {job_id}: {e}")
        errors.append((job_id, str(e)))
```

```python
# Save the results
os.makedirs("cprc_recovery", exist_ok=True)

with open("cprc_recovery/reservoir_states.pkl", "wb") as f:
    pickle.dump(reservoir_states, f)

with open("cprc_recovery/job_ids_used.txt", "w") as f:
    for jid in job_ids_used:
        f.write(jid + "\n")

with open("cprc_recovery/errors.txt", "w") as f:
    for jid, err in errors:
        f.write(f"{jid}: {err}\n")

print(f"\n✅ Completed: {len(reservoir_states)} states saved | {len(errors)} errors logged.")
```
