```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.special import expit as sigmoid
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge, LinearRegression
```

```python
from utils import DataHandler
from ESN import ESNetwork, ClassicalRC
from ESN2 import EchoStateNetwork
from reservoirs import CPRC, GBPermanents
from circuits import CPCircuit
```

```python
window_size = 20
n_samples = 1000
# X, Y = DataHandler().load_dataset('narma', n_samples = 1000, window_size=4, plot=True)
X, Y = DataHandler().load_dataset('mackey_glass', n_samples = n_samples, tau = 25, window_size=window_size,prediction_horizon=20, plot=True)
```

```python
tr = 0.75
train_size = int(n_samples*tr)  # Training set size
# Split into training and testing sets
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = Y[:train_size], Y[train_size:]
```

```python
itr = 50
a= -1.4746251231135765
b = 1.47
vals = np.linspace(a,b,itr)
```

```python
mserror2 =[]
for i in range(itr):
    dim=len(X_train[0])
#     CP_params = [
#                 -np.pi/3, 
#                 np.pi/6, 
#                 -np.pi/9, 
#                 np.pi/7, 
#                 i*np.pi/160, 
#                 -np.pi/7
#             ]
    CP_params = [vals[i], np.pi/7, -np.pi/3, 7*np.pi/7,np.pi/9, -np.pi/7]
    cprc = CPRC(dim=dim, execution_mode='simulation', CP_params=CP_params, kernel = True) #simulation
    esn = ESNetwork(
        reservoir=cprc, 
        dim=dim, 
        regularization=1e-6, 
        alpha=1, 
        show_progress=True, 
        approach='feedback', # 'feedback', or 'time-multiplexing'
        model_type='ridge',  # Options: 'ridge', 'lasso', 'linear', 'svr', 'svc'
        limit = 0.6,
        cpk=True)
    esn.fit(X_train, y_train)
    predictions = esn.predict(X_test)
    rmse = mean_squared_error(y_test, predictions)
    mserror2.append(rmse)
    # print(predictions)
```

```python
# CP_params = [-np.pi/3, i*np.pi/160, -np.pi/9, np.pi/7, np.pi/9, -np.pi/7]
param_values = [i for i in range(len(vals))]
plt.figure(figsize=(8, 5))
plt.plot(param_values, mserror2, marker='o', linestyle='-')
plt.xlabel("1st Parameter in CP_params (radians)")
plt.ylabel("Mean Squared Error")
plt.title("Performance of Quantum Reservoir vs Parameter Sweep")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from qiskit.quantum_info import Statevector, partial_trace, entropy
import matplotlib.pyplot as plt
itr = 50
a= -1.4746251231135765
b = 1.47
vals = np.linspace(a,b,itr)
ent_entropy = []
for i in range(itr):
    # Get the statevector
    d = X_train[1]
    CP_params = [vals[i], np.pi/7, -np.pi/3, 7*np.pi/7,np.pi/9, -np.pi/7]
    qc = CPCircuit(num_features=window_size, reps=1, CP_params=CP_params).CPMap().assign_parameters(d)
    state = Statevector.from_instruction(qc)

    # Total number of qubits
    num_qubits = qc.num_qubits

    # Compute entanglement entropy for subsystems of increasing size
    entropies = []
    labels = []

    for cut_size in range(1, num_qubits):  # Subsystems of size 1 to n-1
        qubits_A = list(range(cut_size))              # Subsystem A
        qubits_B = list(range(cut_size, num_qubits))  # Subsystem B
        rho_A = partial_trace(state, qubits_B)
        S = entropy(rho_A, base=2)
        entropies.append(S)
        labels.append(f"{qubits_A} vs rest")
    ent_entropy.append(entropies)
```

```python
theta_vals = vals
entropies_by_cut = list(zip(*ent_entropy))  # shape: (n-1, len(theta_vals))
plt.figure(figsize=(8, 5))

for cut_idx, entropy_list in enumerate(entropies_by_cut):
    plt.plot(theta_vals, entropy_list, marker='o', label=f'n={cut_idx + 1} vs rest')

plt.plot(theta_vals, np.array(mserror2)*1e5, marker='o', linestyle='-', label = 'MSE')
plt.xlabel("Entangling Parameter (θ = i · π / 160)")
plt.ylabel("Entanglement Entropy")
plt.title("Entanglement Entropy vs Parameter for Different Subsystem Sizes")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
