```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from scipy.special import expit as sigmoid
from sklearn.neural_network import MLPRegressor
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class CustomReservoir:
    def __init__(self, n_reservoir=100, spectral_radius=0.9, activation=np.tanh):
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.activation = activation
    
    def initialize_weights(self, input_dim):
        np.random.seed(42)
        self.Win = np.random.uniform(-1, 1, (self.n_reservoir, input_dim))
        W = np.random.uniform(-1, 1, (self.n_reservoir, self.n_reservoir))
        rhoW = max(abs(np.linalg.eigvals(W)))
        self.W = (W / rhoW) * self.spectral_radius
    
    def compute_state(self, x_prev, u_t):
        return self.activation(np.dot(self.Win, u_t) + np.dot(self.W, x_prev))
```

```python
def generate_narma(n_samples=2000, order=10):
    np.random.seed(42)
    u = np.random.uniform(0, 0.5, n_samples)
    y = np.zeros(n_samples)
    for t in range(order, n_samples - 1):
        y[t + 1] = (0.3 * y[t] + 0.05 * y[t] * np.sum(y[t - order:t]) 
                     + 1.5 * u[t - order] * u[t] + 0.1)
    return u, y

# Define Reservoir Computing (Echo State Network)
class EchoStateNetwork:
    def __init__(self, n_reservoir=100, spectral_radius=0.9, regularization=1e-6):
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.regularization = regularization
    
    def _init_weights(self, input_dim):
        np.random.seed(42)
        self.Win = np.random.uniform(-1, 1, (self.n_reservoir, input_dim))
        W = np.random.uniform(-1, 1, (self.n_reservoir, self.n_reservoir))
        rhoW = max(abs(np.linalg.eigvals(W)))
        self.W = (W / rhoW) * self.spectral_radius
    
    def fit(self, X, Y):
        self._init_weights(X.shape[1])
        
        # Run reservoir
        states = np.zeros((X.shape[0], self.n_reservoir))
        for t in range(1, X.shape[0]):
            states[t] = np.tanh(np.dot(self.Win, X[t]) + np.dot(self.W, states[t - 1]))
        
        # Train readout layer (ridge regression)
        self.Wout = Ridge(alpha=self.regularization).fit(states, Y).coef_.T
    
    def predict(self, X):
        states = np.zeros((X.shape[0], self.n_reservoir))
        for t in range(1, X.shape[0]):
            states[t] = np.tanh(np.dot(self.Win, X[t]) + np.dot(self.W, states[t - 1]))
        return np.dot(states, self.Wout)

# Define RNN and LSTM models
class RNN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(RNN, self).__init__()
        self.rnn = nn.RNN(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        out, _ = self.rnn(x)
        return self.fc(out[:, -1, :])

class LSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(LSTM, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

```

```python
class CustomEchoStateNetwork:
    def __init__(self, n_reservoir=100, spectral_radius=0.9, regularization=1e-6):
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.regularization = regularization
        self.reservoir = CustomReservoir(n_reservoir, spectral_radius)

    def fit(self, X, Y):
        self.reservoir.initialize_weights(X.shape[1])
        
        # Run custom reservoir
        states = np.zeros((X.shape[0], self.n_reservoir))
        for t in range(1, X.shape[0]):
            states[t] = self.reservoir.compute_state(states[t-1], X[t])
#         print("stt:", states)
        # Train readout layer (ridge regression)
        self.Wout = Ridge(alpha=self.regularization).fit(states, Y).coef_.T

    def predict(self, X):
        states = np.zeros((X.shape[0], self.n_reservoir))
        for t in range(1, X.shape[0]):
            states[t] = self.reservoir.compute_state(states[t-1], X[t])
#         print(states)
        return np.dot(states, self.Wout)
```

```python
# Prepare dataset
u, y = generate_narma()
train_size = 1500
test_size = 500

X_train, y_train = u[:train_size].reshape(-1, 1), y[:train_size]
X_test, y_test = u[train_size:].reshape(-1, 1), y[train_size:]

# Train and evaluate models
models = {
    "ESN": EchoStateNetwork(n_reservoir=100),
    "MLP": MLPRegressor(hidden_layer_sizes=(50,), max_iter=1000, random_state=42),
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    results[name] = rmse
    plt.plot(y_test, label="True")
    plt.plot(predictions, label=f"{name} Pred")
    plt.title(f"{name} Predictions")
    plt.legend()
    plt.show()
```

```python
from matrices import OpticalNetwork
from Permanents import RyserPermanent, ClassicalCoincidence
class KernelFunction:
    def __init__(self, network_connections, classical=False, use_angle=True, n_jobs=-1):
        self.network_connections = network_connections
        self.m = len(self.network_connections)
        self.half_m = self.m // 2  # Half the number of beam splitters
        self.classical = classical
        self.use_angle = use_angle
        self.n_jobs = n_jobs

    def encode_input(self, X, epsilon):
        X = np.array(X).flatten()
        if X.shape[0] < self.half_m:
            padding = np.zeros(self.half_m - X.shape[0])  
            encoded_values = np.concatenate((X, padding)) + epsilon
        else:
            encoded_values = X[:self.half_m] + epsilon
        angles = encoded_values if self.use_angle else np.zeros_like(encoded_values)
        phases = np.zeros_like(encoded_values) if self.use_angle else encoded_values
        return angles, phases

    def compute_kernel(self, X, eps=0.001):
        angles_Xi, phi_Xi = self.encode_input(X, epsilon=0)
        angles_Xj, phi_Xj = self.encode_input(X, epsilon=eps)
        new_angles_Xj, new_phi_Xj = np.flip(angles_Xj), np.flip(phi_Xj)
        angles_param = np.concatenate((angles_Xi, new_angles_Xj))
        phases_param = np.concatenate((phi_Xi, new_phi_Xj))
        parameters = list(zip(angles_param, phases_param))
        
        optical_network = OpticalNetwork(self.network_connections, parameters)
        unitary_matrix = optical_network.compute_final_unitary()

        if self.classical:
            kernel_value = ClassicalCoincidence(unitary_matrix, parallel=False, info = False, perm_list=True).calculate()
        else:
            kernel_value = RyserPermanent(parallel=False).compute(unitary_matrix)
        
        return abs(kernel_value)

# class QuantumEchoStateNetwork:
#     def __init__(self, network_connections, regularization=1e-6, classical=False, use_angle=True, n_jobs=-1):
#         self.quantum_reservoir = KernelFunction(network_connections=network_connections, classical=classical, use_angle=use_angle, n_jobs=n_jobs)
#         self.regularization = regularization

#     def fit(self, X, Y):
#         quantum_states = np.array([self.quantum_reservoir.compute_kernel(x) for x in X]).reshape(-1, 1)
#         print(quantum_states)
#         self.Wout = Ridge(alpha=self.regularization).fit(quantum_states, Y).coef_.T

#     def predict(self, X):
#         quantum_states = np.array([self.quantum_reservoir.compute_kernel(x) for x in X]).reshape(-1, 1)
#         print(quantum_states)
#         return np.dot(quantum_states, self.Wout)

class QuantumEchoStateNetwork:
    def __init__(self, network_connections, regularization=1e-6, classical=False, use_angle=True, n_jobs=-1):
        self.quantum_reservoir = KernelFunction(network_connections=network_connections, 
                                                classical=classical, 
                                                use_angle=use_angle, 
                                                n_jobs=n_jobs)
        self.regularization = regularization
        self.model = None  # Store the full Ridge model

    def fit(self, X, Y):
        quantum_states = np.array([self.quantum_reservoir.compute_kernel(x) for x in X])
        
        # Ensure shape consistency
        if quantum_states.ndim == 1:
            quantum_states = quantum_states.reshape(-1, 1)
        
        print("qst:", quantum_states)
        self.model = Ridge(alpha=self.regularization).fit(quantum_states, Y)

    def predict(self, X):
        if self.model is None:
            raise ValueError("Model has not been trained yet. Call `fit` first.")

        quantum_states = np.array([self.quantum_reservoir.compute_kernel(x) for x in X])
        
        # Ensure shape consistency
        if quantum_states.ndim == 1:
            quantum_states = quantum_states.reshape(-1, 1)
        
        print(quantum_states) 
        return self.model.predict(quantum_states)
```

```python
def train_model(model, X_train, y_train, X_test, y_test, epochs=100):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.01)
    
    # Ensure correct shape for PyTorch tensors
    X_train_torch = torch.tensor(X_train, dtype=torch.float32).unsqueeze(-1)
    y_train_torch = torch.tensor(y_train, dtype=torch.float32).unsqueeze(-1)
    X_test_torch = torch.tensor(X_test, dtype=torch.float32).unsqueeze(-1)
    
    for epoch in range(epochs):
        optimizer.zero_grad()
        output = model(X_train_torch)
        loss = criterion(output, y_train_torch)  # Ensure shapes match
        loss.backward()
        optimizer.step()
    
    # Ensure correct shape when returning predictions
    y_pred = model(X_test_torch).detach().numpy().flatten()
    return np.sqrt(mean_squared_error(y_test, y_pred)), y_pred
```

```python
rnn_model = RNN(1, 50, 1)
lstm_model = LSTM(1, 50, 1)

results["RNN"], rnn_pred = train_model(rnn_model, X_train, y_train, X_test, y_test)
results["LSTM"], lstm_pred = train_model(lstm_model, X_train, y_train, X_test, y_test)

# Plot RNN and LSTM predictions
plt.plot(y_test, label="True")
plt.plot(rnn_pred, label="RNN Pred")
plt.title("RNN Predictions")
plt.legend()
plt.show()

plt.plot(y_test, label="True")
plt.plot(lstm_pred, label="LSTM Pred")
plt.title("LSTM Predictions")
plt.legend()
plt.show()

# Compare results
print("Model RMSE Scores:")
for model, rmse in results.items():
    print(f"{model}: {rmse:.4f}")
```
